```python
data = pd.read_csv('/content/healthcare_dataset.csv')

gender_counts = data['Gender'].value_counts()

# Siapkan data untuk pie chart
genders = gender_counts.index
counts = gender_counts.values
colors = ('#DB2E5F', '#62D0DB')
explode = (0, 0)

plt.title('Gender Distribution in Healthcare Dataset')
plt.pie(
    counts,
    labels=genders,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    explode=explode,
    shadow=True
    )

plt.show()
```

Data di atas merupakan visualisasi dari distribusi gender dalam layanan kesehatan. Berdasarkan visualisasi di atas, dapat dilihat bahwa terdapat pie chart yang tergolong dalam dua kelompok yaitu perempuan (Female) berwarna merah muda dan laki-laki (Male) berwarna biru yang dinyatakan dalam persentase(%).

Dari visualisasi data ini dapat disimpulkan bahwa perempuan sedikit lebih banyak berkunjung ke layanan kesehatan dibanding laki-laki dengan persentase 50.7% untuk perempuan dan 49.3% untuk laki-laki. Distribusi gender hampir seimbang dengan selisih 1.4%.
